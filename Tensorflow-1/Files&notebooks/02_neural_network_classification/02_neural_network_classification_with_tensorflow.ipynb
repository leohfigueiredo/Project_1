To visualize our model's predictions, let's create a function which plots the decision boundary of a model in a 2D space.

* Take in a trained model, features and labels
* Create a meshgrid of the different X values
* Make predictions across the meshgrid
* Plot the predictions as well as the original data

```python
import numpy as np
```

```python
def plot_decision_boundary(model, X, y):
    # Define the axis boundaries of the plot
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    
    # Create a prediction
    x_in = np.c_[xx.ravel(), yy.ravel()] # Create a X value
    y_preds = model.predict(x_in) # Make some predictions
    # Check for multi-class or binary classification
    if len(y_preds[0]) > 1:
        print("doing multiclass classification")
        y_preds = np.argmax(y_preds, axis=1).reshape(xx.shape)
    else:
        print("doing binary classification")
        y_preds = np.round(y_preds).reshape(xx.shape)
    
    # Plot the decision boundary
    plt.contourf(xx, yy, y_preds, cmap=plt.cm.RdYlBu, alpha=0.7)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```

# Finding the best learning rate

To find the ideal learning rate (the learning rate where the loss decreases the most during training) we're going to use the following steps:

* A learning rate **callback** â€“ you can think of a callback as an extra piece of functionality you can add to your model *while* it's training.
* Another model (we could use the same one as above, but we're practicing building models here).
* A modified loss curves plot.


```python
# Set random seed
from operator import call


tf.random.set_seed(42)

#Create a model (same as model_8)
model_9 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')   # Output layer for binary
])

# Compile the model
model_9.compile(loss=tf.keras.losses.binary_crossentropy,
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), # changed learning rate to 0.01
                metrics=["accuracy"])

# Create a learning rate callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 *    10**(epoch / 20))

# Fit the model (passing Lr_scheduler callback )
history_9 = model_9.fit(X_train, y_train, epochs=100, callbacks=[lr_scheduler])




```

```python
# Checkout the history
pd.DataFrame(history_9.history).plot(figsize=(12, 6), title="Model_9 Loss curves", xlabel="Epochs")
```

```python
# Plot the learning rate vs loss
lrs = 1e-4 * (10 ** (tf.range(100) / 20))
plt.figure(figsize=(12, 6))
plt.semilogx(lrs, history_9.history["loss"])
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning Rate vs Loss")
plt.grid(True)
plt.show()
```

How about we prettify our confusion matrix?

```python
# Note: the confusion matrix code we're about to write is a remix of scikit-learn's plot_confusion_matrix
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.plot_confusion_matrix.html?highlight=plot%20confusion%20matrix"

import itertools
import numpy as np

figsize = (10, 10)

# Create the confusion matrix

cm = confusion_matrix(y_test,  tf.round(y_preds)) #round to 0 or 1,y_preds closer to 0 = 0, closer to 1 = 1, cm = confusion_matrix(y_test, y_preds)
cm_norm = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis] # normalize the confusion matrix
n_classes = cm.shape[0] # number of classes we're dealing with

# lets prettify it
fig, ax = plt.subplots(figsize=figsize)
# Create a matrix plot
cax = ax.matshow(cm, cmap=plt.cm.Blues) # color map = plt.cm.Blues
fig.colorbar(cax)

# Create classes
classes = False

if classes:
    labels = classes
else:
    labels = np.arange(cm.shape[0])

# Label the axes
ax.set(title="Confusion Matrix",
      xlabel="Predicted label",
      ylabel="True label",
      xticks=np.arange(n_classes),
      yticks=np.arange(n_classes),
      xticklabels=labels,
      yticklabels=labels)

# Set x-axis labels to bottom
ax.xaxis.set_label_position("bottom")
ax.xaxis.tick_bottom()

#Adjust the label size
ax.xaxis.label.set_size(20)
ax.yaxis.label.set_size(20)
ax.title.set_size(20)

# Set the threshold for different colors
threshold = (cm.max() + cm.min()) / 2.

# Plot the text on the plot
# Plot the text on the plot
for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, f"{cm[i, j]}\n({cm_norm[i, j]*100:.1f}%)",
             horizontalalignment='center',
             color="white" if cm[i, j] > threshold else "black",
             size=15,
             va="center",
             ha="center")
plt.show()
```

## Building a multi-class classification model

For our multiclass classification example, we're going to use a similar architecture to the one we used for our binary classification example.

* Input shape = 28x28(the shape of a single image)
* Output shape = 10 (the number of classes in the Fashion MNIST dataset)
* Loss function = sparse categorical crossentropy (this is the multiclass classification version of binary crossentropy) tf.keras.losses.SparseCategoricalCrossentropy()
    * Note: We use sparse categorical crossentropy when our target labels are integers. If our target labels were one-hot encoded, we'd use categorical crossentropy tf.keras.losses.CategoricalCrossentropy()
* Output layer activation = softmax (this is the multiclass classification version of sigmoid) tf.keras.activations.softmax

```python
len(class_names)
```

```python
# Our data needs to be flattened before being fed into a neural network
flatten_model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28))])  # Flatten the 2D images into 1D array
flatten_model.output_shape  # returns (None, 784) 28 * 28 = 784
```

```python
# Set the random seed
tf.random.set_seed(42) 

# 1. Create a model
model_11 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),  # Flatten the input images
    tf.keras.layers.Dense(4, activation='relu'),     
    tf.keras.layers.Dense(4, activation='relu'),    # hidden layer
    tf.keras.layers.Dense(10, activation='softmax') # output layer (10 classes) 
])

# 2. Compile the model
model_11.compile(loss=tf.keras.losses.CategoricalCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

# 3. Fit the model
non_norm_history = model_11.fit(train_data, tf.one_hot(train_labels, depth=10), epochs=10, validation_data=(test_data, tf.one_hot(test_labels, depth=10)))
# one-hot encode the labels (turn them into one-hot vectors) since we're dealing with a multi-class classification problem, depth=10 for 10 different classes, 
```

## Evaluating our multiclass classification model

To evaluate our multiclass classification model we could:

- Evaluate its performance using other classification metrics (such as a confusion matrix) 
- Assess some of its predictions (through visualizations)
- Improve its results (by training it for longer or changing the architecture)
- Save and export it for use in an application


```python
def make_confusion_matrix(y_true, y_pred, classes=None, figsize=(10, 10), text_size=15, norm=False, savefig=False):
    # Create the confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    if norm:
        cm_norm = cm.astype("float") / cm.sum(axis=1, keepdims=True)
    else:
        cm_norm = cm
    n_classes = cm.shape[0]

    # Prettify the plot
    fig, ax = plt.subplots(figsize=figsize)
    cax = ax.matshow(cm, cmap=plt.cm.Blues)
    fig.colorbar(cax)

    # Set labels to the correct classes if passed in
    labels = classes if classes is not None else np.arange(n_classes)

    # Label the axes
    ax.set(
        title="Confusion Matrix",
        xlabel="Predicted label",
        ylabel="True label",
        xticks=np.arange(n_classes),
        yticks=np.arange(n_classes),
        xticklabels=labels,
        yticklabels=labels
    )

    # Set x-axis labels to bottom
    ax.xaxis.set_label_position("bottom")
    ax.xaxis.tick_bottom()

    # Adjust the label size
    ax.xaxis.label.set_size(text_size)
    ax.yaxis.label.set_size(text_size)
    ax.title.set_size(text_size)

    # Set the threshold for different colors
    threshold = (cm.max() + cm.min()) / 2.

    # Plot the text on the plot
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        value = cm[i, j]
        percent = cm_norm[i, j]*100 if norm else (cm[i, j]/cm.sum(axis=1)[i]*100 if cm.sum(axis=1)[i] > 0 else 0)
        plt.text(
            j, i, f"{value}\n({percent:.1f}%)",
            horizontalalignment='center',
            color="white" if value > threshold else "black",
            va="center",
            ha="center",
            fontsize=text_size
        )
    if savefig:
        plt.savefig("confusion_matrix.png")
    plt.show()              
```

ðŸ”‘ Note:
Often when working with images and other forms of visual data, it's a good idea to visualize as much as possible to develop a further understanding of the data and the inputs and outputs of your models.
How about we create a fun little function for:
- Plot a random image
- Make a prediction on said image
- Label the plot with the truth label & the predicted label


```python
import random

def plot_random_image(model, images, true_labels, classes):
    """
    Plots a random image along with its true and predicted labels.
    Args:
        model: A trained TensorFlow/Keras model.
        images: A numpy array of images to choose from.
        true_labels: The true labels corresponding to the images.
        classes: A list of class names corresponding to label indices.
    """
    i = random.choice(range(len(images)))  # Randomly select an index
    target_image = images[i]  # Select the image at the random index
    pred_probs = model.predict(target_image.reshape(1, 28, 28))  # Predict the label for the selected image
    pred_label = classes[(pred_probs.argmax())]  # Get the predicted class name
    true_label = classes[true_labels[i]]  # Get the true class name

    # Plot the image
    plt.imshow(target_image, cmap=plt.cm.binary)
    
    # Change the color of the title based on whether the prediction is correct
    if pred_label == true_label:
        color = "green"
    else:
        color = "red"

    #Add xlabels information (predicted label, true label)
    plt.xlabel("Pred:{} {:2.0f}% (True: {})".format(pred_label,
                                                     100 * tf.reduce_max(pred_probs),
                                                     true_label),color=color)  # Change the color of the xlabel based on prediction correctness
    plt.axis(False)
    plt.show()
```

```python
import random

#  Function to plot a random image along with its true and predicted labels

def plot_random_image(model, images, true_labels, classes):
    import matplotlib.pyplot as plt
    i = random.choice(range(len(images)))
    target_image = images[i]
    pred_probs = model.predict(target_image.reshape(1, 28, 28))
    pred_label = classes[pred_probs.argmax()]
    true_label = classes[true_labels[i]]
    plt.imshow(target_image, cmap=plt.cm.binary)
    color = "green" if pred_label == true_label else "red"
    plt.title(f"Pred: {pred_label} ({100 * tf.reduce_max(pred_probs):.0f}%)\nTrue: {true_label}", color=color, fontsize=16)
    plt.axis(False)
    plt.show()
```
